```python
with open('/content/S12.txt') as f:
    lines = f.readlines()

print(lines[0])
words = lines[0].split()
print(lines[1])
print(lines[2])
print(lines[3])
print(lines[4])
print(words)
print(words[0])
```

```python
ld1 = 0.0; ld4 = 0.0; dx = 0.0; currentf=0; currents=0;
l={'Frequency':[],'S12':[],'ld1':[],'ld4':[],'dx':[]}
for count,line in enumerate(lines):

  if count%1004==0:
    print(f'changing {count}')
    a=line.split()
    #print(a, type(a))
    ld1 = a[24]
    ld1 = float(ld1[4:-1])
    ld4 = a[27]
    ld4 = float(ld4[4:-1])
    dx = a[14]
    dx = float(dx[3:-1])

  elif count%1004>=3:
    #print(count)
    f,s = line.split()
    l['Frequency'].append(f)
    l['S12'].append(s)
    l['ld1'].append(ld1)
    l['ld4'].append(ld4)
    l['dx'].append(dx)

df = pd.DataFrame.from_dict(l)
df
```
